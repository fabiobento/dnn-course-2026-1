## Pré-requisitos

### Para entender
* **Python** – você não precisa ser um programador especialista em Python, mas precisa conhecer o básico. Se não conhece, o [tutorial oficial do Python](https://docs.python.org/3/tutorial/) é um bom lugar para começar.
* **Python científico** – Usaremos algumas bibliotecas populares do Python, em particular NumPy, matplotlib e pandas. Se você não estiver familiarizado com essas bibliotecas, provavelmente deve começar pelos tutoriais na seção Ferramentas (especialmente NumPy).
* **Matemática** – Também usaremos algumas noções de álgebra linear, cálculo, estatística e teoria da probabilidade. Você deve ser capaz de acompanhar se já aprendeu esses conceitos no passado, pois não serão muito avançados, mas se não conhece esses tópicos ou precisa de uma atualização, consulte a introdução apropriada na seção Matemática.

### Para executar os exemplos
* **Jupyter** – Esses notebooks são baseados no Jupyter. Você pode executá-los com apenas um clique usando o Google Colab (ou outra plataforma, como Kaggle, Binder, Deepnote ou Colaboratory): não é necessária instalação. Você também pode apenas visualizá-los usando o visualizador do Jupyter.org ou instalar tudo em seu computador, conforme preferir.
